```python
import pandas as pd 
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import joblib
from sklearn import tree
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn import metrics
```

```python
df=pd.read_csv('music.csv')
df
```

```python
x=df.drop(columns=['genre'])  
y=df['genre']
x_train,x_test,y_train ,y_test=train_test_split(x,y,test_size=.2)
model=DecisionTreeClassifier()
model.fit(x,y)
```

```python
print('DecisionTreeClassifierModel Train Score is : ' , model.score(x_train, y_train))
print('DecisionTreeClassifierModel Test Score is : ' , model.score(x_test, y_test))
```

```python
y_pred=model.predict(x_test)
predictions=model.predict([[20,1]])
predictions
```

```python
print("precision_score = ",metrics.precision_score(y_test, y_pred,average='micro'))
```

```python
print("Recall_score= ",metrics.recall_score(y_test, y_pred,average='micro'))
```

```python
print("F1-Score = ",metrics.f1_score(y_test, y_pred,average='micro'))
```

```python
joblib.dump(model,'music_recommender.joblib')
y_pred=model.predict(x_test)
score =accuracy_score(y_test,y_pred)
print('accuracy_score=',score)
```

```python
metrics.confusion_matrix(y_test,y_pred)
metrics.plot_confusion_matrix(model.fit(x,y),x,y,['HipHop','Jazz','Classical','Acoustic','Dance'])
```

```python
fig, axes = plt.subplots(nrows = 1,ncols = 1,figsize = (4,4), dpi=300)

tree.plot_tree(model,
           feature_names =  ['age','gender'], 
           class_names='genre',
           filled = True);
```

```python
importance = model.feature_importances_
# summarize feature 
for i,v in enumerate(importance):
	print('Feature: %0d, Score: %.5f' % (i,v))
plt.bar([x for x in range(len(importance))], importance)
plt.show()
```
